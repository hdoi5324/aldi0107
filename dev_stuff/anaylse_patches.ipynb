```python
import os
from glob import glob

import pandas as pd

patch_dir = "/home/heather/phd_data/loose/loose_nosam_17711/artifacts"
csv_files = glob(os.path.join(patch_dir, "*.csv"))
n = 100
data = [pd.read_csv(f, index_col=False).T for f in csv_files[:n]]
df = pd.concat(data).reset_index()
print(df.index)
```

```python
df.columns = [1, 2, 3, 4, 'N1', 'N2', 'Pos', 'Neg']
df['file'] = [os.path.basename(f) for f in csv_files[:n]]
df = df.sort_values(by=['Pos'])
```

```python
print(df[-30:])
```

```python
from IPython.display import Image, display, HTML

for index, row in df.tail(20).iterrows():
    # Display image
    print(row)
    filename = row['file']
    display(Image(os.path(patch_dir, f"{filename[:-3]}jpg")))
    
    # Display caption
    display(HTML(f"<figcaption style='text-align:center; font-style:italic;'>{filename}</figcaption>"))
```
